```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Consider Entire Population:

```python
# To consider the Entire population, we have to take both the 'train' and 'test' dataset.


train_df = pd.read_csv('titanic_train.csv')
test_df = pd.read_csv('titanic_test.csv')
```

```python
df = pd.concat([train_df.drop(columns = ['Survived']), test_df])
df
```

### Original Population Distribution:

```python
df['Fare'].plot(kind = 'kde')
```

### Shuffle all the rows:

```python
df = df.sample(1309).dropna()
df
```

### Take 50 Random Samples & repaet for 100 times: 

```python
samples = []

for i in range(100):
    samples.append(df['Fare'].sample(50).values.tolist())
```

```python
samples
```

### Converting samples to array:

```python
samples = np.array(samples)
samples
```

```python
samples.shape
```

### Sample Mean:

```python
sample_mean = np.mean(samples, axis = 1)
sample_mean
```

### Plot:

```python
sns.kdeplot(sample_mean)
```

```python
sns.kdeplot(sample_mean)
plt.axvline(df['Fare'].mean(), c = 'r', ls = '--')
```
